```python
import pandas as pd # 판다스를 불러온다
```

```python
# 각 열들의 데이터를 생성한다.

titles = ['Skyfall', 'Lose Yourself', 'Another Brick in the Wall', 'Use Somebody', 'Treasure']
Artists = ['Adele', 'Eminem', 'Pink Floyd', 'Kings of Leon', 'Bruno Mars']
Lengths = ['4:46', '5:26', '3:59', '3:51', '2:58']
Years = [2012, 2002, 1979, 2008, 2013]
```

```python
# 각 항목 별로 위에 있는 벡터를 넣어주는 데이터 프레임을 df에 저장한다.

df = pd.DataFrame({'Title':titles, 'Artist':Artists, 'Length':Lengths, 'Year':Years})
```

```python
df
```

```python
# title에 해당하는 값들을 소문자로 바꿔서 리스트의 형태로 출력한다. (내 방법은 안좋은 방법)

title_lowercase = [df.Title[0].lower(),df.Title[1].lower(),df.Title[2].lower(),df.Title[3].lower(),df.Title[4].lower()]
```

```python
title_lowercase
```

```python
# 위에 문제에 대한 조교님 솔루션

title_lowercase = [t.lower() for t in df['Title']]
```
